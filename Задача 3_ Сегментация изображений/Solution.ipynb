```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
import os
import torch
```

```python
targets = list(os.listdir('targets'))
images = list(os.listdir('images'))
```

```python
for image in range(len(images)):
    img = cv.imread(f'images/{images[image]}', cv.IMREAD_GRAYSCALE)
    assert img is not None, "file could not be read, check with os.path.exists()"
    #img = cv.medianBlur(img,3)
    ret, th = cv.threshold(img, 
                             0, 255,  
                             cv.THRESH_BINARY + cv.THRESH_OTSU) 
    imgs = [img, th]
    titles = ['Original Image', 
                'Thresholding']
    for i in range(2):
        plt.subplot(2,2,i+1),plt.imshow(imgs[i],'gray')
        plt.title(titles[i])
        plt.xticks([]),plt.yticks([])
    plt.show()
```
